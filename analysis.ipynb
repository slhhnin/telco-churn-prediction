```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
df = pd.read_csv('Dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Clean TotalCharges
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
```

```python

# 2. Inspect Churn Distribution (Highlighting Class Imbalance)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Distribution of Customer Churn')
plt.show()
```

```python
# 3. Analyze Key Correlates (e.g., Internet Service vs. Churn)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='InternetService', hue='Churn', palette='viridis')
plt.title('Churn Rate by Internet Service Type')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Print out specific insights
print(pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100)
```
